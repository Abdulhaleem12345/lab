```python

import pandas as pd

# Load the dataset
df = pd.read_excel("/content/EVEN (1).xlsx", sheet_name="Sheet1")

# 1. Average total cost of trips for each destination
avg_cost_per_destination = df.groupby("Destination")["Total_Cost"].mean()
print("Average Total Cost per Destination:\n", avg_cost_per_destination)

# 2. Distribution of payment methods
payment_distribution = df["Payment_Method"].value_counts()
print("\nPayment Method Distribution:\n", payment_distribution)

# 3. Number of unique destinations
unique_destinations = df["Destination"].nunique()
print("\nNumber of Unique Destinations:", unique_destinations)

# 4. Create a new column for trip duration
# Assuming Travel_Dates are in the format "start_date - end_date"
def calculate_duration(travel_dates):
    try:
        start_date, end_date = travel_dates.split(" - ")
        start_date = pd.to_datetime(start_date.strip(), errors='coerce')
        end_date = pd.to_datetime(end_date.strip(), errors='coerce')
        return (end_date - start_date).days if pd.notnull(start_date) and pd.notnull(end_date) else None
    except:
        return None

df["Trip_Duration"] = df["Travel_Dates"].apply(calculate_duration)
print("\nSample Data with Trip Duration:\n", df[["Travel_Dates", "Trip_Duration"].head()])

# 5. Check for missing values in key columns
missing_values = df[["Travel_Package", "Payment_Method", "Travel_Agent"]].isnull().sum()
print("\nMissing Values in Important Columns:\n", missing_values)
```
